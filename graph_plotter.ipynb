```python
from graph_analysis import create_graph_from_files
import networkx as nx
import matplotlib.pyplot as plt
```

```python
as_number = "18429"
G = nx.Graph()
create_graph_from_files(as_number, G, "data_2020_08", links_type="fm")
largest_cc = max(nx.connected_components(G), key=len)
G_lcc = G.subgraph(largest_cc)
nx.draw(G_lcc, node_size=25)
plt.show()
```
